```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Public_transport = 'Updated-Public-transport.csv'
df = pd.read_csv(Public_transport)
df
```

```python
df = df.drop(df.index[0:11])
df
```

```python
df = df.iloc[:, [0,1,2,3,4,5]]
df
```

In the following plot, I couldn't get the values in the y-axis in the right order. Do you have any idea to fix this?

```python
first_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 1e 20%-groep']
second_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 2e 20%-groep']
third_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 3e 20%-groep']
fourth_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 4e 20%-groep']
fifth_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 5e 20%-groep']

plt.plot(first_group_data.columns[1:], first_group_data.values[0][1:], label='First income group', marker='o')
plt.plot(second_group_data.columns[1:], second_group_data.values[0][1:], label='Second income group', marker='o')
plt.plot(third_group_data.columns[1:], third_group_data.values[0][1:], label='Third income group', marker='o')
plt.plot(fourth_group_data.columns[1:],fourth_group_data.values[0][1:], label='Fourth income group', marker='o')
plt.plot(fifth_group_data.columns[1:], fifth_group_data.values[0][1:], label='Fifth income group', marker='o')

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Participation in public transport over the years for different income groups')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```
