```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = 'data/ovop.csv'
df = pd.read_csv(csv_file_path)
df
```

```python
# df.set_index('Perioden', inplace=True)



ax = df.T.plot(kind='line', marker='o', figsize=(10, 6))

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Participation in public transport over the years for different age and gender groups')

ax.legend(title='Rij', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid()
plt.show()
```
