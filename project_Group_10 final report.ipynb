# Research Objective

By the end of 2019, the COVID-19 pandemic had spread over the globe and had a profound effect on every area of peoples' daily lives. Many steps were taken to restrict social connections and provide adequate social distance in order to stop the virus from spreading. As a result, daily travel habits have drastically altered. In order to get a deeper understanding of the reasons behind the changes in travel patterns during a worldwide crisis regarding he usage of public transport in The Netherlands, we classified individuals into different groups. These groups are based on differences in gender, income and age. As a result, we compared and analysed the travel patterns of these groups prior to and during the COVID-19 pandemic. Our goal is to obtain an overview of the effects of COVID-19 on travel behaviour based on these analyses.

**Main objective:**

What effect did Covid-19 have on the usage of public transport in the Netherlands for different groups?

**Sub-objectives:**

- What effect did Covid-19 have on the usage of public transport in the Netherlands for different gender groups?

- What effect did Covid-19 have on the usage of public transport in the Netherlands for different income groups?

- What effect did Covid-19 have on the usage of public transport in the Netherlands for different age groups for both man and women?

Answering these questions should help us better understand the influence the pandamic and its restrictions had on the usage of public transport in The Netherlands.

**Subquestion 1**: What effect did Covid-19 have on the usage of public transport in the Netherlands for different gender groups?

```python
import csv  # Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = "data/ptdata2.csv" # Path to the datafile
df = pd.read_csv(dataset) # Defining the data
print(df) # Creating a preview of the data table
```

```python
# Filter the DataFrame to get data for 'Male' and 'Female' separately
male_data = df[df['Periods'] == 'Male'] # Select rows where the index is 'Male'
female_data = df[df['Periods'] == 'Female'] # Select rows where the index is 'Female'

plt.plot(male_data.columns[1:], male_data.values[0][1:], label='Male', marker='o') # Create a line plot for 'Male' data
plt.plot(female_data.columns[1:], female_data.values[0][1:], label='Female', marker='o') # Create a line plot for 'Female' data

plt.xlabel('Year') # Label the x-axis
plt.ylabel('Percentage') # Label the y-axis
plt.title('Participation in public transport over the years for male and female') # Set a title for the plot

plt.legend() # Display a legend to differentiate 'Male' and 'Female' lines
plt.grid() # Display a grid in the plot
plt.show() # Show the plot


```

The line graph above shows the participation in public transport over the five year period, for both men and women. First of all, the differences between men and women do not appear to be very big, both lines show a steady decline from 2019 to 2020. Furthermore, after 2020 the participation seems to steadily grow again. 

When it comes to the differences between men and women, then it appears that women have a slightly higher participation compared to men for all years. However, this appears to be a small difference. Participation for women was around 9% in 2018, and stayed the same for 2019. In 2020, the participation dropped by more than half, remaining at 4.2%. In 2021, this participation grew to 4.7%, and grew even more to 6.5% in 2023. For men, participation was around 8% in both 2018 and 2019. This dropped to 3.6% in 2020. The participation grew in a similar trend as the women, with a participation of 3.8% in 2021, and 5.6% in 2022. 

```python
# Rewrite data to make it usable for the heatmap
data = {
    'Periods': ['Male', 'Female'],
    2018: [8.0, 9.2],
    2019: [7.9, 9.2],
    2020: [3.6, 4.2],
    2021: [3.8, 4.7],
    2022: [5.6, 6.5]
}

df = pd.DataFrame(data) # Create a DataFrame from the data
df = df.drop(columns=['Periods']) # Remove the 'Periods' column as it contains non-numeric data

plt.figure(figsize=(12, 9)) # Setting the size for the heatmap
heatmap = sns.heatmap(df, cmap='OrRd', annot=True, fmt='.1f') # Creating the heatmap with different colours of red, including the values in the map

heatmap.set_yticklabels(['Male', 'Female']) # Set y-axis labels to 'Male' and 'Female'

plt.title('Heatmap for participation in public transport over the years for men and women') # Adding a title
plt.xlabel('Years') # Adding title for x-axis
plt.ylabel('Gender') # Adding title for y-axis

colourbar = heatmap.collections[0].colorbar # Adding the colourbar with title and size
colourbar.set_label('Percentage of participants', fontsize=15) # Adding a label to the colourbar and a font size

plt.show() # Printing the heatmap
```

The data is alternatively displayed in the heatmap above. Using this heatmap, it becomes even more clear how big the difference is between 2019 and 2020. This shows the impact of Covid measures on public transport significantly impacted the participation in public transport, for both men and women (Centraal Bureau voor de Statistiek, 2023). It can also be noted that the participation from before Covid has not fully recovered yet in 2022, although the participation is on an upward trend.

**Subquestion 2**: What effect did Covid-19 have on the usage of public transport in the Netherlands for different income groups?

```python
Public_transport = 'data/p3.csv'
df = pd.read_csv(Public_transport)
df
```

```python
df = df.drop(df.index[0:11])
df
```

```python
df = df.iloc[:, [0,1,2,3,4,5]]
df
```

```python
first_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 1e 20%-groep']
second_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 2e 20%-groep']
third_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 3e 20%-groep']
fourth_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 4e 20%-groep']
fifth_group_data = df[df['Perioden'] == 'Gestandaardiseerd inkomen: 5e 20%-groep']

plt.plot(first_group_data.columns[1:], first_group_data.values[0][1:].astype(float), label='First income group', marker='o')
plt.plot(second_group_data.columns[1:], second_group_data.values[0][1:].astype(float), label='Second income group', marker='o')
plt.plot(third_group_data.columns[1:], third_group_data.values[0][1:].astype(float), label='Third income group', marker='o')
plt.plot(fourth_group_data.columns[1:],fourth_group_data.values[0][1:].astype(float), label='Fourth income group', marker='o')
plt.plot(fifth_group_data.columns[1:], fifth_group_data.values[0][1:].astype(float), label='Fifth income group', marker='o')

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Participation in public transport over the years for different income groups')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```

All groups were affected by Covid-19 regarding the use of public transport. As seen in the plot, all have a drop in 2020 when Covid started. Then in 2021, only the first two income groups started to use public transport more. The other groups stayed the same or even decreased a little bit. Eventually in 2022, the percentage of all groups increased and is slowly growing towards pre-Covid.

With the exception of the first income group, all groups are such close with the percentages that the differences are not significant enough to be distinguished. Only the first group is significant different. They originally used public transport a lot more. This resulted in a bigger drop in percentage when Covid hit in comparison to the other groups. But the percentage of the use also increased more in 2021 and 2022.

**Subquestion 3**: What effect did Covid-19 have on the usage of public transport in the Netherlands for different age groups for both man and women?

```python
csv_file_path = 'data/ovopp2.csv' #path to the datafile
df = pd.read_csv(csv_file_path) 
print(df.head()) #printing a preview of the data table

df.set_index('Age/gender', inplace=True) #setting age/gender as the index, so that 0-13 is not taken into account
ax = df.T.plot(kind='line', marker='o', figsize=(10, 6)) #creating a line graph with circles at datapoints and defining the figuresize the 
plt.xlabel('Year') #setting (axis) titles
plt.ylabel('Percentage')
plt.title('Participation in public transport over the years for different age and gender groups')

ax.legend(title='Line for age and gender', bbox_to_anchor=(1.05, 1), loc='upper left') #adding a legend and defining the title at location

plt.grid() #adding a grid for easier reading 
plt.show()

plt.figure(figsize=(12, 9)) #Setting the size for the heatmap
heatmap = sns.heatmap(df, cmap='OrRd', annot=True, fmt='.1f') #creating the heatmap with different colours of red, including the values in the map
plt.title('Heatmap for participation in public transport over the years for different age and gender groups') #adding a title
plt.xlabel('Years') #adding title for x-axis
plt.ylabel('Age and gender') #adding title for y-axis
colourbar = heatmap.collections[0].colorbar #adding the colourbar with title and size
colourbar.set_label('Percentage of participants', fontsize=15)
plt.show() #printing the heatmap
```

This first graph, the linegraph, shows us the participations of different agegroups for both man and women in public transport. For all the different groups, we see a drop in use of public transport for the year 2020, presumably due to the covid restrictions. The biggest drop in percentage is for man and women of age 18-25. However, the group of men in agerage 25-35 have a relatively higher loss of participation, their percentage is more than halved, from 13.8 to 5.8%. Also for women in the same agegroup, there is an big drop in the usage. 

The headmap clearly shows us that the younger generations make in percentage better use of the public transport and are thusly most effected by the restrictions. It also shows that the public transport participation is not back at the levels of before Covid, but that it is gaining in. 
In 2020 public transport was at its lowest low due to Covid, then in the following years the public transport started to gain little be little again. It would be interesting to at the year 2023 to this research as soon as the data is available as this is the first year completely without any restrictions. Then we can see if the grow of the last two years in continued.
