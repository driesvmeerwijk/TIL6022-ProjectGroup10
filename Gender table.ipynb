```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file_path = "C:\\Users\\sande\\Desktop\\ptdata2.csv"
df = pd.read_csv(csv_file_path)
print(df)
```

```python
male_data = df[df['Periods'] == 'Male']
female_data = df[df['Periods'] == 'Female']

plt.plot(male_data.columns[1:], male_data.values[0][1:], label='Male', marker='o')
plt.plot(female_data.columns[1:], female_data.values[0][1:], label='Female', marker='o')

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Participation in public transport over the years for male and female')

plt.legend()
plt.grid()
plt.show()


```
